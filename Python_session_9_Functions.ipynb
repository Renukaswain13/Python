```python
a=eval(input("enter the value of a:"))
b=eval(input("enter the value of b:"))
def add(a,b):
    print(a+b)
    
def sub(a,b):
    print(b-a)
    
def mul(a,b):
    print(a*b)
    
def div(a,b):
    print(a/b)

def calculator():
    option=eval(input("enter the number between 1-4:"))
    if option==1:
        add(a,b)
    elif option==2:
        sub(a,b)
    elif option==3:
        mul(a,b)
    elif option==4:
        div(a,b)

calculator()
    
```
